Token Embdedding with embedding layer

```python
import torch
import torch.nn as nn
import tiktoken
```

```python
text1="My name is Asmina Nassar"
text2= "i am awsome"
text = "<|endoftext|>".join([text1,text2])

tokenizer= tiktoken.get_encoding('gpt2')
encoded_id= torch.tensor(tokenizer.encode(text, allowed_special={"<|endoftext|>"}))
print(encoded_id)
```

```python
vocabulary_size = tokenizer.n_vocab
output_dim = 5
```

```python
embedding=nn.Embedding(vocabulary_size,output_dim)
```

```python
text_test = "Asmina is awsome"
tokenizer_new= tiktoken.get_encoding('gpt2')

encoded_id_new=torch.tensor(tokenizer_new.encode(text_test,disallowed_special=()))
print(encoded_id_new)
embedding(encoded_id_new)
```
