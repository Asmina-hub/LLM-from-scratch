```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
!pip install PyPDF2
```

```python
import PyPDF2
with (open('harrypotter.pdf','rb')) as f:
  reader = PyPDF2.PdfReader(f)
  pdf_text = ""
  for i in range(11,len(reader.pages)-1):
    page = reader.pages[i]
    pdf_text = pdf_text + page.extract_text()

print(f"Total extracted characters: {len(pdf_text)}")
print("First 500 characters of the extracted text:")
print(pdf_text[:500])
```

```python
preprocessing = re.split(r'([,.:;?_!"()\'-]|--|\s)' , pdf_text)
preprocessing =[i.strip() for i in preprocessing if i.strip()]
print(f"Total tokens: {len(preprocessing)}")
print(preprocessing[:50])
```

```python
vocab_word = sorted(set(preprocessing))
vocab_word.extend(["<|endoftext|>", "<|unk|>"])
vocab = {}
reverse_vocab = {}
for integer, word in enumerate(vocab_word):
  vocab[word] = integer
  reverse_vocab[integer] = word
print(len(vocab), len(reverse_vocab))
```

```python
class SimpleTokenizerV1:
  def __init__(self, vocab,reverse_vocab):
    self.vocab = vocab
    self.reverse_vocab = reverse_vocab

  def encode(self,text):
    preprocessing = re.split(r'([,.:;?_!"()\'-]|--|\s)' , text)
    print('preprocessing done')
    preprocessing =[i.strip() for i in preprocessing if i.strip()]
    preprocessed = [ i if i in self.vocab else "<|unk|>" for i in preprocessing ]
    ids =[self.vocab[i] for i in preprocessed] # Changed from 'preprocessing' to 'preprocessed'
    print(len(ids))
    return ids

  def decode(self,ids):
    text = " ".join( [self.reverse_vocab[i] for i in ids])
    text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
    return text
```

```python
tokenizer = SimpleTokenizerV1(vocab,reverse_vocab)

text = "At half past eight, Mr. Dursley picked up his briefcase, pecked Mrs.Dursley on the cheek"
ids = tokenizer.encode(text)
print(ids)
```

```python
tokenizer = SimpleTokenizerV1(vocab,reverse_vocab)
text = tokenizer.decode(ids)
print(text)
```

What you observed above is the out of vocabulary error

```python
tokenizer = SimpleTokenizerV1(vocab,reverse_vocab)

text1 = "At half past eight, Mr. Dursley picked up his briefcase,"
text2 = "pecked Mrs.Dursley on the cheek"
text = " <|endoftext|> ".join((text1, text2))
ids = tokenizer.encode(text)
print(ids)
```

```python
tokenizer = SimpleTokenizerV1(vocab,reverse_vocab)
text = tokenizer.decode(ids)
print(text)
```
